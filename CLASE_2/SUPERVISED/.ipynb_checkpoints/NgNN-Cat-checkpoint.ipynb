### Modulos en Python.


<font color='black'>

Un módulo es un archivo que contiene definiciones y declaraciones en Python destinado para su uso en otros programas de Python.

Hay muchos módulos de Python que vienen con Python como parte de la biblioteca estándar.

Para usar un módulo de Python se debe importar:

<font color='green'> $\textbf{import}$   <font color='black'> name_module
    
o
    
<font color='green'> $\textbf{from}$   <font color='black'>name_module <font color='green'> $\textbf{import}$   <font color='purple'> *

<font color='black'>
    
Con la declaración anterior se tiene acceso a todas las definiciones dentro del modulo name_module.

Si solo se quiere tener acceso a ciertas deficiniones, se usa lo siguiente:

<font color='green'> $\textbf{from}$   <font color='black'>name_module <font color='green'> $\textbf{import}$   <font color='black'>name_function



Además de los modulos definidos en la libreria de Python se pueden crear modulos propios que son archivos .py que contienen definiciones y declaraciones en python. Se importan de la misma manera. Por ejemplo, para importar el archivo dnn_app_utils_v3.py el cual contiene funciones que definidas por el usuario (se omite el .py):




<font color='green'> $\textbf{from}$   <font color='black'> dnn_app_utils_v3 <font color='green'> $\textbf{import}$   <font color='pruple'> *
    
    
Los modulos se pueden importar con cualquiern nombre:

<font color='green'> $\textbf{import}$   <font color='black'>  name_module <font color='green'> $\textbf{as}$   <font color='pruple'> alias

#### Se crea el modelo de la red neuronal:


<img src="nlayer.png" style="width:750px;height:500px;">



```python

def L_layer_model(X, Y, layers_dims, learning_rate = 0.0075, num_iterations = 3000, print_cost=False):#lr was 0.009
    """
    Implements a L-layer neural network.
    
    Arguments:
    
    X -- data, numpy array of shape (number of examples, num_px * num_px * 3)
    
    Y -- true "label" vector (containing 0 if cat, 1 if non-cat), of shape (1, number of examples)
    
    layers_dims -- list containing the input size and each layer size, of length (number of layers + 1).
    
    learning_rate -- learning rate of the gradient descent update rule
    
    num_iterations -- number of iterations of the optimization loop
    
    print_cost -- if True, it prints the cost every 100 steps
    
    Returns:
    
    parameters -- parameters learnt by the model. They can then be used to predict.
    """
    
    tic = time.process_time()
    
    # Devuelve el valor (en fracciones de segundo) de la suma del sistema y el tiempo de CPU del usuario del proceso 
    # actual.

    np.random.seed(1)
    
    # Los numeros aleatorios se generan a partir de la semilla 1. Esto permite obtener resultados reproducibles, ya que 
    # siempre se generaran los mismo valores aleatorios. 
    
    costs = []                         
    
    # se crea una lista vacía
    
    parameters = initialize_parameters_deep(layers_dims)
    
    # initialize_parameters_deep ---> inicializa a los parametros w y b.
    
    for i in range(0, num_iterations):

        # Forward propagation: [LINEAR -> RELU]*(L-1) -> LINEAR -> SIGMOID.
        AL, caches = L_model_forward(X, parameters)
        
        # Compute cost.
        cost = compute_cost(AL, Y)
    
        # Backward propagation.
        grads = L_model_backward(AL, Y, caches)
 
        # Update parameters.
        parameters = update_parameters(parameters, grads, learning_rate)
                
        
        if print_cost and i % 100 == 0:
            print ("Cost after iteration %i: %f" %(i, cost))
            costs.append(cost)
            
            # si print_cost = True e i es divisible por 100 ---> Se imprime el costo y el valor dado se guarda en la
            # lista cost.
            
    pred_train = predict(train_x, train_y, parameters)
    
    toc = time.process_time()
    
    print ( "\n ----- Computation time = " + str(1000*(toc - tic)) + "ms")

            
    # plot the cost
    plt.plot(np.squeeze(costs))
    plt.ylabel('cost')
    plt.xlabel('iterations')
    plt.title("Learning rate =" + str(learning_rate))
    plt.show()
    
    # se grafica el costo por iteración. 
    
    return parameters


# Las funciones:

    # initialize_parameters_deep

    # L_model_forward

    # compute_cost

    # L_model_backward

    # update parameters
    
# se definen en dnn_app_utils_v3.py
```

#### La arquitectura de la red nos permite elegir el numero de capas intermedias así como el numero de unidades por capa, esta información se indica en el arreglo layers_dims.

```python
layers_dims = [12288, 20, 7, 5, 3, 1]

# se define una lista en donde el valor de cada entrada indica el numero de unidades que tendra la red en esa capa:

    # Input tiene 12288 unidades

    # La primera capa intermedia tiene 20 unidades

    # La segunda capa intermedia tiene 7 unidades

    # La tercera capa intermedia tiene 5 unidades

    # La cuarta capa intermedia tiene 3 unidades

    # la capa de salida tiene una unidad
    
# En total la red definida con la lista layer_dims tiene 5 capas (La primer capa (input) no se cuenta).

parameters = L_layer_model(train_x, train_y, layers_dims, num_iterations = 2500, print_cost = True)
```
